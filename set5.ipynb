```python
# 1. Write a sorting algorithm (e.g., bubble sort) and test its functionality.
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

test_list = [64, 34, 25, 12, 22, 11, 90]
print("Original list:", test_list)
print("Sorted list:", bubble_sort(test_list))
```

```python
# 2. Create a program that utilizes a hashtable to store and retrieve key-value pairs.
class CustomHashTable:
    def __init__(self, size=100):
        self.size = size
        self.table = [[] for _ in range(self.size)]
    
    def _hash(self, key):
        return hash(key) % self.size
    
    def insert(self, key, value):
        hash_index = self._hash(key)
        for item in self.table[hash_index]:
            if item[0] == key:
                item[1] = value
                return
        self.table[hash_index].append([key, value])
    
    def get(self, key):
        hash_index = self._hash(key)
        for item in self.table[hash_index]:
            if item[0] == key:
                return item[1]
        raise KeyError(key)

ht = CustomHashTable()
ht.insert("name", "Alice")
ht.insert("age", 30)
print("Name:", ht.get("name"))
print("Age:", ht.get("age"))
```

```python
# 3. Develop a function that sorts a list of dictionaries based on the values of a specific key.
def sort_dictionaries_by_key(dict_list, key):
    return sorted(dict_list, key=lambda x: x[key])

employees = [
    {"name": "Alice", "age": 30, "salary": 50000},
    {"name": "Bob", "age": 25, "salary": 45000},
    {"name": "Charlie", "age": 35, "salary": 60000}
]

print("Sorted by age:")
age_sorted = sort_dictionaries_by_key(employees, "age")
for emp in age_sorted:
    print(emp)

print("\nSorted by salary:")
salary_sorted = sort_dictionaries_by_key(employees, "salary")
for emp in salary_sorted:
    print(emp)
```

```python
# 4. Write a program that implements a breadth-first search for a graph structure.
from collections import defaultdict, deque

class Graph:
    def __init__(self):
        self.graph = defaultdict(list)
    
    def add_edge(self, u, v):
        self.graph[u].append(v)
    
    def bfs(self, start):
        visited = set()
        queue = deque([start])
        visited.add(start)
        bfs_traversal = []
        
        while queue:
            vertex = queue.popleft()
            bfs_traversal.append(vertex)
            
            for neighbor in self.graph[vertex]:
                if neighbor not in visited:
                    queue.append(neighbor)
                    visited.add(neighbor)
        
        return bfs_traversal

g = Graph()
g.add_edge(0, 1)
g.add_edge(0, 2)
g.add_edge(1, 2)
g.add_edge(2, 0)
g.add_edge(2, 3)
g.add_edge(3, 3)

print("BFS starting from vertex 2:")
print(g.bfs(2))
```

```python
# 5. Use Matplotlib to plot a simple line graph representing a mathematical function.

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = x**2  # Quadratic function

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='y = x')
plt.title('Line Graph of Quadratic Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 6. Create a program that plots a scatter plot for random data points.
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)  # For reproducibility
num_points = 100
x = np.random.randn(num_points)
y = np.random.randn(num_points)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, c='blue', alpha=0.6, edgecolors='black')
plt.title('Scatter Plot of Random Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
# 7. Implement a bar chart visualizing sales data over a month.
import matplotlib.pyplot as plt

weeks = ['Week 1', 'Week 2', 'Week 3', 'Week 4']
sales = [15000, 22000, 18500, 25000]

plt.figure(figsize=(10, 6))
plt.bar(weeks, sales, color='green', edgecolor='black')
plt.title('Monthly Sales by Week')
plt.xlabel('Weeks')
plt.ylabel('Sales Amount ($)')

for i, v in enumerate(sales):
    plt.text(i, v + 500, f'${v}', ha='center')

plt.tight_layout()
plt.show()
```
